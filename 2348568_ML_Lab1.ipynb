```python
cd=churn_data1[['Gender','Geography','HasCrCard',	'IsActiveMember','Exited']]
cd
```

```python
from sklearn.preprocessing import LabelEncoder
for col in cd.columns:
  le=LabelEncoder()
  cd[col]=le.fit_transform(cd[col])
```

```python
cd.head()
```

```python
from sklearn.feature_selection import chi2
X=cd.drop(columns=['Exited'])
Y=cd['Exited']
```

```python
chi_scores=chi2(X,Y)
chi_scores
```

```python
chi_values=pd.Series(chi_scores[0],index=X.columns)
chi_values.sort_values(ascending=False,inplace=True)
chi_values.plot.bar()
```

```python
from scipy.stats import chi2_contingency
chi2, p_value, dof, expected  = chi2_contingency(chi_scores)
print(f"Chi-Square Value: {chi2}")
print(f"P-Value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies:{expected}")
```

```python
sig_lev= 0.05
if p_value <= sig_lev:
    print('We Reject Null hypothesis: there is a significant relationship between isActiveMember, Gender , Geography and Exited')
else:
    print('We fail to reject null hypothesis; there is no significant relationship between isActiveMember, Gender, Geography and Exited')
```

# **ADDITIONAL TASK OF RESAMPLING AND SYNTHETIC DATA GENERATION**

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.combine import SMOTEENN
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
```

```python
X_tr, X_test, Y_tr, Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)

classifier = RandomForestClassifier(random_state=42)
```

```python
smote = SMOTE(random_state=42)
resampled_X, resampled_Y = smote.fit_resample(X_tr, Y_tr)
```

```python
classifier.fit(resampled_X, resampled_Y)
```

```python
Y_predict = classifier.predict(X_test)
```

```python
RUS = RandomUnderSampler(random_state=42)
RUS_resampled_X, RUS_resampled_Y = RUS.fit_resample(X_tr, Y_tr)
```

```python
classifier.fit(RUS_resampled_X,RUS_resampled_Y )
```

```python
RUS_predict_Y = classifier.predict(X_test)
```

```python
print("Classification Report (Random UnderSampler):")
print(classification_report(Y_test,RUS_predict_Y ))
```

```python
SE = SMOTEENN(random_state=42)
SE_resampled_X, SE_resampled_Y= SE.fit_resample(X_tr, Y_tr)
```

```python
classifier.fit(SE_resampled_X,SE_resampled_Y )
```

```python
print("Classification Report (SMOTEENN):")
print(classification_report(Y_test, RUS_predict_Y ))
```
